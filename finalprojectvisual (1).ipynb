```python
import csv
from matplotlib import pyplot as plt
```

```python
filename = 'populkz.csv'

with open(filename) as f:
    reader = csv.reader(f)
    header_row = next(reader)

    population, years = [], []
    for row in reader:
        number = str(row[1])
        population.append(number)
        year = str(row[0])
        years.append(year)

new_years = years[::5]
new_years = [int(x) for x in new_years]
new_population = population[::5]
new_population = [x.replace(',', '') for x in new_population]
new_population = [int(x) for x in new_population]
```

```python
plt.style.use('seaborn')
fig, ax = plt.subplots(figsize=(15,6))

ax.plot(new_years, new_population, c = 'red')
ax.scatter(new_years, new_population, c = 'blue', cmap = plt.cm.Blues, edgecolors = 'none', s = 90)

plt.title('Annual growth of the KZ population', fontsize = 24)
plt.xlabel('Years', fontsize=25)
plt.ylabel('Population', fontsize=25)
plt.tick_params(axis='both', which = 'major', labelsize = 10)
plt.ticklabel_format(style = 'plain')

plt.show()
```

```python
import csv
from plotly.graph_objs import Bar, Layout
from plotly import offline
```

```python
filename = 'populkz.csv'

with open(filename) as f:
    reader = csv.reader(f)
    header_row = next(reader)

    growth_percent, years = [], []
    for row in reader:
        percent = str(row[2])
        growth_percent.append(percent)
        year = str(row[0])
        years.append(year)
```

```python
new_growth = growth_percent[::5]
new_growth = [x.replace('N/A', '0') for x in new_growth]
new_growth = [x.replace(' %', '') for x in new_growth]
new_growth = [float(x) for x in new_growth]

new_years = years[::5]
new_years = [int(x) for x in new_years]

```
